```python
### 定義清理資料函式
def clean_data(df1, output_file, targetcol, head, branch, method, IPC, IPCcol):
    replacedict = dict(zip(head, branch))
    if method == "normal":
        df = df1.set_index(df1.columns.drop(targetcol, errors='ignore').tolist())[targetcol].str.split('|', expand=True).stack().reset_index().rename(columns={0: targetcol})
        df[targetcol] = df[targetcol].str.lstrip()
        df[targetcol] = df[targetcol].str.rstrip()
        df[targetcol] = df[targetcol].str.rstrip(';')
        df[targetcol] = df[targetcol].str.split(n=1).str[0]
        df = df.dropna(subset=[targetcol])
        df = df[df[targetcol].str.endswith(('公司', '企業'))]
        df[targetcol] = df[targetcol].str.replace('（', '(')
        df[targetcol] = df[targetcol].str.replace('）', ')')
        df[targetcol] = df[targetcol].str.replace('Ｆ', 'F')
        df[targetcol] = df[targetcol].str.replace('Ｔ', 'T')
        df[targetcol] = df[targetcol].str.replace('Ｃ', 'C')
        # 公司名稱替換
        df[targetcol] = df[targetcol].replace(replacedict)
    elif method == "copatent":
        df = pd.read_excel(df1)
        df[targetcol] = df[targetcol].str.lstrip()
        df[targetcol] = df[targetcol].str.rstrip()
        df[targetcol] = df[targetcol].str.rstrip(';')
        df[targetcol] = df[targetcol].str.split(n=1).str[0]
        df = df.dropna(subset=[targetcol])
        df = df[df[targetcol].str.endswith(('公司', '企業'))]
        df[targetcol] = df[targetcol].str.replace('（', '(')
        df[targetcol] = df[targetcol].str.replace('）', ')')
        df[targetcol] = df[targetcol].str.replace('Ｆ', 'F')
        df[targetcol] = df[targetcol].str.replace('Ｔ', 'T')
        df[targetcol] = df[targetcol].str.replace('Ｃ', 'C')
        # 刪除分公司名稱(分公司共同專利若計入會引響總公司，因分公司的專利跟組公司一起申請)
        for key in branch:
            df = df[~df.apply(lambda row: key in row.to_string(), axis=1)]
    else:
        print("方法指定錯誤")
        
    if IPC == 'yes':
        df = df.set_index(df.columns.drop(IPCcol, errors='ignore').tolist())[IPCcol].str.split('|', expand=True).stack().reset_index().rename(columns={0: IPCcol})
        df[IPCcol] = df[IPCcol].str.lstrip()
    else:
        df = df    
        
    df.to_excel(output_file + '.xlsx', index=False)
    print("資料已清理至" + output_file + '\n接下來將會進行資料計算……')
```

```python
### 定義統計資料函式
def calculate(file_name, firmcol, yearcol, targetcol, method, contenttype, direction_file):
    # 讀取Excel檔案
    target = pd.read_excel(file_name)
    
    ## 統計資料
    if method == "count":
        # 在DataFrame中添加一個新的欄位，將公司和年份組合成一個新的欄位
        target[firmcol+yearcol] = target[firmcol] + '_' + target[yearcol].astype(str)
        
        # 使用樞紐分析計算每家公司在每年份中出現的次數(即專利數量，使用count計算出現次數)
        pivot = target.pivot_table(index=firmcol, columns=yearcol, values=firmcol+yearcol, aggfunc='count', fill_value=0)
    
    elif method == "sum":
        # 將專利被引用數量轉換為整數
        target[targetcol] = target[targetcol].astype(int)
        
        # 使用樞紐分析計算每家公司在每年的專利被引用數量(使用sum將每份專利的被引用數量加總)
        pivot = target.pivot_table(index=firmcol, columns=yearcol, values=targetcol, aggfunc='sum', fill_value=0)
        
    elif method == "mean":
        # 將專利被引用數量轉換為整數
        target[targetcol] = target[targetcol].astype(int)
        
        # 使用樞紐分析計算每家公司在每年的專利被引用數量(使用sum將每份專利的被引用數量加總)
        pivot = target.pivot_table(index=firmcol, columns=yearcol, values=targetcol, aggfunc='mean', fill_value=5)
        
    else:
        print("對象指定錯誤")
    
    ## 補充所有年份及公司名稱
    direction = pd.read_excel(direction_file)
    
    # 指定要顯示的年份範圍（1990年到2023年）
    year_range = range(1990, 2023)

    # 獲取所有在資料中出現的公司名單
    all_companies = direction[firmcol].unique()
    
    if method == "mean":
        # 將所有在資料中出現的公司添加到索引中(因為不是所有公司都有進入，此步驟是讓沒進入的公司也出現在表格中，以便後續分析)
        pivot = pivot.reindex(all_companies, fill_value=5)
        
        # 重新索引以包含所有範圍內的年份(同上，不是所有年份都有公司進入，所以要加入所有研究過程的年分)
        pivot = pivot.reindex(columns=year_range, fill_value=5)
        
    else:
        # 將所有在資料中出現的公司添加到索引中(因為不是所有公司都有進入，此步驟是讓沒進入的公司也出現在表格中，以便後續分析)
        pivot = pivot.reindex(all_companies, fill_value=0)
        
        # 重新索引以包含所有範圍內的年份(同上，不是所有年份都有公司進入，所以要加入所有研究過程的年分)
        pivot = pivot.reindex(columns=year_range, fill_value=0)

    # 將結果寫入Excel檔案
    output_file = file + '_' + contenttype + '.xlsx'
    pivot.to_excel(output_file)
    print(file + contenttype + "已統計至" + output_file)
```

```python
def extract_IPC(data_frame, outputfile):
    # 定義要查找的重要IPC列表
    target_strings = ['H02P01', 'H02P03', 'H02P05', 'H02P06', 'H02P07', 'H02P09', 'H02P21', 'H02P25', 'H02P27', 'H02P29', 
                      'H02P31', 'H02K17', 'H02K19', 'H02K21', 'H02K23', 'H02K25', 'H02K27', 'H02K29', 'H02K41', 'H01M02', 
                      'H01M04/13', 'H01M04/14', 'H01M04/58', 'H01M04/82', 'H01M06/14', 'H01M06/16', 'H01M08', 'H01M10', 
                      'B60K06', 'B60L07', 'B60L11', 'B60L15', 'B60T08/17', 'G60W10', 'G60W20', 'H01M10/44', 'H01M10/46', 
                      'H02J07', 'B60L03', 'H02J', 'G01R19', 'G01R13', 'G01R31/02', 'G01R31/04', 'G01R31/06', 'G01R31/07', 
                      'G01R31/36']

    # 創建一個布林索引，找到包含目標字符串的行
    mask = data_frame['Orig. IPC(All)'].str.contains('|'.join(target_strings), case=False, na=False)

    # 使用布林索引過濾數據
    filtered_IPC = data_frame[mask]

    filtered_IPC.to_excel(outputfile + '.xlsx', index=False)
```

```python
def extract_copatent(input_file, targetcol, output_file):
    # 讀取 Excel 檔案
    df = input_file
    
    # 選擇包含 | 的行
    filtered_df = df[df[targetcol].str.contains("\|")]

    # 保存新的 Excel 文件
    filtered_df.to_excel(output_file + '.xlsx', index=False)
```

```python
def delete_files(file_list):
    for file_name in file_list:
        os.remove(file_name)
```

```python
def merge(files, contenttype, firmcol, method):
    dfs = {}

    for i, file in enumerate(files, 1):
        data = str(file+'_'+contenttype+'.xlsx')
        df_name = f'df{i}'
        df = pd.read_excel(data)
        dfs[df_name] = df

    merged = pd.concat(list(dfs.values()), ignore_index=True)
    if method == 'mean':
        final = merged.groupby(firmcol).mean().reset_index()
        final.to_excel(contenttype + '(合併).xlsx', index=False)
    else:
        final = merged.groupby(firmcol).sum().reset_index()
        final.to_excel(contenttype + '(合併).xlsx', index=False)
    print(contenttype + '已合併至 ' + contenttype + '(合併).xlsx')
```

```python
def delete_pre_merge(files, contenttypes):
    del_all = []
    for file,contenttype in zip(files, contenttypes):
        del_quant = str(file+'_' + contenttype + '.xlsx')
        del_all.extend([del_quant])

    for delete in del_all:
        os.remove(delete)
    print('原資料已刪除')
```

```python
def convert(input_file, firmcol, firms, years, contenttype):
    # 讀取Excel文件
    df = pd.read_excel(input_file)

    # 使用melt函數將寬格式轉為長格式
    conv = df.melt(id_vars=[firmcol], var_name='年分', value_name='數量')

    # 重新排序列
    conv = conv[[firmcol, '年分', '數量']]

    # 重新命名
    conv.columns = [firms, years, contenttype]

    # 保存为新的Excel文件
    conv.to_excel(contenttype + '轉換' + '.xlsx', index=False)
    
    print(contenttype+'已轉換至 '+contenttype + '轉換' + '.xlsx')
```

```python
def merge_convent(contenttypes, output_file):
    file_names = []
    for contenttype in contenttypes:
        file_name = contenttype + '轉換' + '.xlsx'
        file_names.append(file_name)
        
    # 讀取第一个文件
    df1 = pd.read_excel(file_names[0])

    # 合併剩餘的文件
    for file_name in file_names[1:]:
        df_temp = pd.read_excel(file_name)
        column_to_insert = df_temp.iloc[:, 2]
        df1 = pd.concat([df1, column_to_insert], axis=1)

    # 储存合併後的文件
    op = output_file + '.xlsx'
    df1.to_excel(op, index=False)
    print('轉換後文件已合併至 ' + op)
```

```python
### 定義計算共同專利百分比的函式
def calculate_percentage(input_file, outputfile):
    # 读取Excel文件
    df = pd.read_excel(input_file)

    # 计算百分比并将结果存储在新列中
    df['co_percent'] = (df['copatent'] / df['quant'])
    
    # 空值補零
    df['co_percent'].fillna(0, inplace=True)
    
    # 將修改後的 DataFrame 寫回 Excel 檔案
    op = outputfile + '.xlsx'
    df.to_excel(op, index=False)
    
    print('共同專利百分比已加入至' + op)
```

```python
def move_RD(inputfile, outputfile):
    # 讀取Excel文件
    df = pd.read_excel(inputfile)
    
    # 將每一行的資料向後移動兩個位置
    df.iloc[:, 2:] = df.iloc[:, :-2].values
    
    # 將"1990"列和"1991"列的值全部設為零
    df.iloc[:, 1:3] = 0
    
    # 將結果保存到一個新的Excel文件
    df.to_excel(outputfile + '.xlsx', index=False)
```

```python
def change_name(inputfile, contractfile, origcol, contcol, outputfile):
    # 假設你已經有一個 DataFrame
    df = pd.read_excel(inputfile)
    df3 = pd.read_excel(contractfile)

    # 迭代每一行的 'ser' 列
    for index, row_ser in df.iterrows():
        # 檢查 'ser' 行是否為有效字串，避免處理缺失值
        if pd.notna(row_ser[origcol]):
            # 迭代 'fal' 列中的每一個元素
            for fal_value in df3[contcol]:
                # 檢查 'ser' 行的值是否是 'fal' 列中的任一元素的子字串
                if row_ser[origcol].strip() in fal_value.strip():
                    df.at[index, origcol] = fal_value
                    break  # 如果找到匹配，就停止迭代 'fal' 列

    # 將合併後的結果寫入新的 Excel 檔
    df.to_excel(outputfile + '.xlsx', index=False)
```

```python
def insert(inputfile, convertfile, firms, years, targetcol, outputfile):
    # 讀取兩個Excel文件
    df1 = pd.read_excel(inputfile)
    df2 = pd.read_excel(convertfile)
    
    # 將"公司行"和"年份"列作為索引
    df1.set_index([firms, years], inplace=True)
    df2.set_index([firms, years], inplace=True)
    
    # 將df1的"RD"列加入到df2中
    df2[targetcol] = df1[targetcol]
    
    # 將索引還原為列
    df1.reset_index(inplace=True)
    df2.reset_index(inplace=True)
    
    ##### 刪除為空值的行 #####
    df2 = df2.dropna(subset=[targetcol])
    
    # 將 rd 行的缺失值設置為0
    df2[targetcol].fillna(0, inplace=True)
    
    # 將結果保存到一個新的Excel文件
    op = outputfile + '.xlsx'
    df2.to_excel(op, index=False)
    print('研究經費已加入至' + op)
```

```python
def age(inputfile, targetcol, namecol, outputfile):
    df = pd.read_excel(inputfile)
    
    # 第一步驟：提取年份
    df[targetcol] = df[targetcol].astype(str).str[:4]
    
    # 第二步驟：創建1990~2022的列名
    years = list(range(1990, 2023))
    df_panel = pd.DataFrame(columns=years)
    
    # 第三步驟：計算存續年分
    for index, row in df.iterrows():
        company_years = df_panel.columns.astype(int) - int(row[targetcol])
        df_panel = pd.concat([df_panel, pd.DataFrame([company_years], columns=df_panel.columns)], ignore_index=True)

    # 將非零的(負數)改為零
    df_panel[df_panel < 0] = 0
    
    # 在新的 Excel 檔案的第一列添加公司名稱
    df_panel.insert(0, namecol, df[namecol])
    
    # 將結果寫入新的 Excel 檔案
    result_excel_file_path = outputfile + '.xlsx'  # 替換為你想要的結果 Excel 檔案路徑
    df_panel.to_excel(result_excel_file_path, index=False)
    
```

```python
import numpy as np

def distance(inputfile, target, outputfile):
    df = pd.read_excel(inputfile)
    
    for index, row in df.iterrows():
        targetcol = row[target]
        
        # 检查是否为缺失值
        if pd.notna(targetcol):
            # IPC 行的判断
            if targetcol[0] in ['H', 'G', 'B']:
                if targetcol[0] == 'H':
                    if targetcol[1:3] == '01':
                        if len(targetcol) > 4 and targetcol[4] == 'M':
                            df.at[index, 'distance'] = 0
                        else:
                            df.at[index, 'distance'] = 2
                    elif targetcol[1:3] == '02':
                        if len(targetcol) > 4 and targetcol[4] in ['P', 'K', 'J']:
                            df.at[index, 'distance'] = 0
                        else:
                            df.at[index, 'distance'] = 2
                    else:
                        df.at[index, 'distance'] = 4
                elif targetcol[0] == 'G':
                    if targetcol[1:3] == '01':
                        if len(targetcol) > 4 and targetcol[4] == 'R':
                            df.at[index, 'distance'] = 0
                        else:
                            df.at[index, 'distance'] = 2
                    elif targetcol[1:3] == '60':
                        if len(targetcol) > 4 and targetcol[4] == 'W':
                            df.at[index, 'distance'] = 0
                        else:
                            df.at[index, 'distance'] = 2
                    else:
                        df.at[index, 'distance'] = 4
                elif targetcol[0] == 'B':
                    if targetcol[1:3] == '60':
                        if len(targetcol) > 4 and targetcol[4] in ['L', 'K', 'T']:
                            df.at[index, 'distance'] = 0
                        else:
                            df.at[index, 'distance'] = 2
                    else:
                        df.at[index, 'distance'] = 4
            else:
                df.at[index, 'distance'] = 5
        else:
            # 处理缺失值的逻辑，这里将 distance 列设置为 np.nan
            df.at[index, 'distance'] = 5
    
    # 将结果保存到 Excel 文件
    op = outputfile + '.xlsx'
    df.to_excel(op, index=False)
```
